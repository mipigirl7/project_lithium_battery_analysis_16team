```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
import platform
```

```python
df = pd.read_csv("metadata.csv")
test_df = df.copy()
```

```python
# 중복된 값과 그 횟수 확인
top_duplicated = test_df['Capacity'].value_counts()
print(top_duplicated[top_duplicated > 1].head(10))
```

```python
# Capacity가 중복된 모든 행 추출 (원본 데이터 확인)
duplicated_rows = test_df[test_df.duplicated(subset=['Capacity'], keep=False)]

# 값이 같은 것끼리 모아서 정렬해보기
duplicated_rows.sort_values(by='Capacity').head(10)
```

```python
# 문자열을 숫자로 변환 (변환할 수 없는 값은 NaN 처리)
test_df['Capacity_numeric'] = pd.to_numeric(test_df['Capacity'], errors='coerce')

# 변환 후 다시 unique 확인
print(f"숫자 변환 전 고유값: {test_df['Capacity'].nunique()}")
print(f"숫자 변환 후 고유값: {test_df['Capacity_numeric'].nunique()}")
```

```python
# 숫자 변환 전(문자열)의 고유값 리스트 추출
unique_str = set(test_df['Capacity'].dropna().unique())

# 숫자 변환 후의 고유값을 다시 문자열로 바꿔서 비교 준비
# (비교를 위해 다시 str로 변환하되, 동일한 포맷으로 통일)
test_df['temp_str'] = test_df['Capacity_numeric'].astype(str)
unique_num_as_str = set(test_df['temp_str'].unique())

# 문자열일 때는 달랐는데 숫자로 바꾸니 같아진 값들 찾기
# '2.5'와 '2.50' 같은 관계를 가진 후보들을 출력합니다.
print("숫자로 변환되면서 하나로 합쳐진 후보들:")
print(test_df.groupby('Capacity_numeric')['Capacity'].unique()[lambda x: x.map(len) > 1])
```

```python
# Capacity 컬럼에서 숫자가 아닌 것들만 골라내기
# (숫자로 변환했을 때 NaN이 되는 행들 찾기)
non_numeric_capacity = test_df[pd.to_numeric(test_df['Capacity'], errors='coerce').isna()]['Capacity'].unique()

print("숫자가 아니어서 NaN으로 변해버린 값들:")
print(non_numeric_capacity)
```
