```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 설정
batteries = ['B0029', 'B0030', 'B0031', 'B0032']
colors = ['red', 'blue', 'green', 'orange']
labels = ['B0029 (2.0V)', 'B0030 (2.2V)', 'B0031 (2.5V)', 'B0032 (2.7V)']

plt.figure(figsize=(12, 7))

# 3. 각 배터리별로 루프를 돌며 데이터 추출 및 그래프 생성
for i, battery in enumerate(batteries):
    # 해당 배터리의 방전(discharge) 데이터만 필터링
    subset = test_df[(test_df['battery_id'] == battery) & (test_df['type'] == 'discharge')]
    
    # 그래프 그리기 (x축은 단순히 실행 순서인 사이클 번호)
    # plt.plot(range(len(y_values)), y_values, label=labels[i], color=colors[i], alpha=0.8)
    # 만약 index가 0부터 시작하지 않을 수 있으므로 .values를 사용해 리스트처럼 취급합니다.
    y_values = subset['Capacity'].values
            
    # 그래프 그리기 (x축은 단순히 실행 순서인 사이클 번호)
    plt.plot(range(len(y_values)), y_values, label=labels[i], color=colors[i], alpha=0.8)

# 4. 그래프 디테일 설정
plt.title('Capacity Degradation Comparison (at 43°C)', fontsize=16)
plt.xlabel('Cycle Number', fontsize=12)
plt.ylabel('Capacity (Ahr)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

EOL 도달 사이클 계산 코드
코드로 정확히 "몇 번째 사이클에서 죽었는지" 숫자를 뽑기

```python
for battery in batteries:
    subset = test_df[(test_df['battery_id'] == battery) & (test_df['type'] == 'discharge')]
    
    initial_cap = subset['Capacity'].iloc[0]
    threshold = initial_cap * 0.7
    
    # threshold보다 작아지는 첫 번째 지점 찾기
    eol_cycle = subset[subset['Capacity'] <= threshold]
    
    if not eol_cycle.empty:
        # eol_cycle의 인덱스나 순번을 가져옴
        cycle_num = len(subset[subset['Capacity'] > threshold])
        print(f"{battery}: {cycle_num}번째 사이클에서 EOL 도달")
    else:
        print(f"{battery}: 아직 EOL에 도달하지 않음 (현재 최저 용량: {subset['Capacity'].min():.4f})")
```
